## Business Questions

1. Why do customers leave the company?

2. Which customers are most likely to churn?

3. What customer characteristics are associated with churn?

4. Which customer segments have the highest churn rate?

5. How much revenue is at risk due to customer churn?

6. What actions can be taken to reduce churn?

7. Which customers should be prioritized for retention efforts?

## Success Metrics

### Business Metrics

- Churn Rate
- Retention Rate
- Revenue Saved
- Customer Lifetime Value

### Machine Learning Metrics

- Precision
- Recall
- F1 Score
- ROC-AUC Score

## Assumptions

- Historical customer behavior can help predict future churn.
- Customer characteristics influence churn decisions.
- Retention actions can reduce customer churn.
- The dataset represents actual customer behavior.
